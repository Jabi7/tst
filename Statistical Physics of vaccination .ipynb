## SIR Model

The SIR eqautions are,
1. $S' = -\beta SI$
2. $I' =  \beta SI - \gamma I$
3. $R = 1 - S - I $ 

where, $S$ is the fraction of susceptible population, $I$ is the fraction of infected population, $R$ is recovered population. 

Equation 2 can also be written as,

4. $I' =  \gamma I(R_0S - 1)$

where, $R_0$ is the ''basic reproduction number'' (BRN) defined as,

5. $R_0 = \beta D$

where $D$ is the expected duration of the infective phase ($D = 1/\gamma$ ).
The maximum of the prevalence can be determined by the curve $I(S)$, which is obtained from Eqs. (1)–(4) by taking,

6. $dI/dS = 1/(R_0S) - 1$

whose solution in $0 < S ≤ S(0)$ (remind that $S(t)$ is decreasing) is,

7. $I(S) = I(0) + S(0) - S + log(S/S(0))/R_0$

 


```python
# defining I(S)
def I_S(s, r0, i0, s0):
    Is = i0 + s0 - s + log(s/s0)/r0
    return Is
```

```python
x  = np.linspace(0, 1, 1000)
s = x[1:]
v = np.vectorize(I_S) 
a = v(s, 5, 0, 1)
b = v(s, 3, 0, 1)
c = v(s, 2, 0, 1)
plt.ylim([0, 0.5])
plt.plot(s, a, label = '$R_0$ = 5')
plt.plot(s, b, label = '$R_0$ = 3')
plt.plot(s, c, label = '$R_0$ = 2')
plt.xlabel("S(t)")
plt.ylabel("I(t)")
plt.title("dynamics in the S ,I phase plane")
plt.legend()
```
